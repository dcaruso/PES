```python
from __future__ import division
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('img/mandrill.tiff',1)
img = cv2.cvtColor(img, cv2.cv.CV_BGR2RGB) # Matplotlib espera RGB
plt.imshow(img)

```

```python
# Acceso a pixel
px = img [20,180]
print px # B, G, R
```

```python
# Acceso a ROI
roi = img[30:80,150:200]
plt.imshow(roi)
```

```python
# Split and merge channels
r,g,b = cv2.split(roi)
plt.imshow(r,cmap='gray')
```

```python
b[:,:]=0 # Hace 0 todos los pixeles azules
img_r = cv2.merge((r,g,b))
plt.imshow(img_r)
```

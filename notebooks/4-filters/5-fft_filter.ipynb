```python
import scipy.signal as sig
import numpy as np
from scipy.fftpack import fft, ifft, fftshift, rfft, irfft
from __future__ import division
from IPython.html.widgets import *
import matplotlib.pyplot as plt

def fft_filter(signal,*bin_to_delete):
    spec = (fft(signal))
    for i in bin_to_delete:
        spec[i]=0
        spec[len(spec)-i]=0
    return ifft(spec)
    
N = 1024
Fs = 10000
Fc1 = 1000
Fc2 = 2000
t = np.arange(0, N/Fs, 1/Fs)
x1 = 10* np.sin(2*np.pi*Fc1*t)
x2 = 10* np.sin(2*np.pi*Fc2*t)
s = x1+x2
plt.plot(abs(fft(s)))
plt.figure()
plt.plot(s[0:100])
```

```python
a = np.arange(190,220,1)
sf = fft_filter(s,a)
plt.plot(abs(fft(sf)))
plt.figure()
plt.plot(sf[0:100])
```

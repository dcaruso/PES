# Medición de SNR y calculo de ENOB

El siguiente notebook tiene por finalidad ayudar en la comprensión del cálculo de SNR (Signal to Noise Ratio) y ENOB (Effective Number Of Bits).
En primer lugar cabe destacar que el ENOB es una forma de hablar sobre el SNR, sólo que el ENOB es una medición del SNR sobre un ADC ideal que sólo introduce ruido de cuantización. Entonces al medir el SNR de un ADC cualquiera y calcular el ENOB, estamos comparando nuestro ADC con uno ideal y determinamos cuantos de esos bits son utilizables.

## Procedimiento:

* Se realiza la captura de señal senoidal con el ADC
* Se intenta buscar la señal seoindal mas parecida a las muestras capturadas por el ADC (fitting)
* La señal "fiteada" se usa como señal pura (su energía será el numerador del SNR)
* La diferencia entre la señal "fiteada" y la medida con el ADC será el ruido.
* Se calcula SNR y ENOB a partir de los resultados

Para más detalles, ver: http://www.mit.edu/~klund/A2Dtesting.pdf
